## notebook identifies carnitine compounds in ALL_GNPS library by annotation description

```python
import pandas as pd
```

```python
# from v_get_ALL_GNPS_input_library.ipynb

input_library_full_df_loaded = pd.read_parquet('/home/jovyan/work/notebooks/outputs/ALL_GNPS_input_library.gzip')
```

```python
input_library_full_df_loaded.columns
```

```python
input_library_no_suspect_list = input_library_full_df_loaded[~input_library_full_df_loaded['library_membership'].str.contains('suspect',case=False,na=False)]
```

```python
adduct_labels = ['M+H','[M+H]','[M+H]+']
```

```python
input_library = input_library_no_suspect_list[input_library_no_suspect_list["Adduct"].isin(adduct_labels)]
```

```python
len(input_library)
```

## shaping data

```python
# filtered data to use for analysis only containing information for M+H adduct carnitines
input_library_of_carnitines_case_insen_MpH = input_library[input_library['Compound_Name'].str.contains('carnitine',case=False,na=False)]
```

```python
input_library_of_carnitines_case_insen_MpH.reset_index().to_csv(
    '/home/jovyan/work/notebooks/outputs/library_df_carnitine_case_insen_M+H.csv', 
    sep=',', index=False)
```
